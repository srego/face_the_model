```python
import numpy as np
import pandas as pd
import keras
import glob
from collections import defaultdict
from keras.preprocessing import image
```

```python
model = '/home/ubuntu/h_hour/test_1/test_1.h5'
paul = '/home/ubuntu/h_hour/paul_holdout'
actors = '/home/ubuntu/h_hour/test_1/actors_copy'
actresses = '/home/ubuntu/h_hour/test_1/actresses_copy'
cats = '/home/ubuntu/h_hour/cats'
dogs = '/home/ubuntu/h_hour/dogs'
shape = (256, 256)
```

```python
classifier = keras.models.load_model(model)

def paul_dict_test(directory):
    # Make sure path is in correct format. 
    if directory[-1] != '/':
        directory += '/'
    
    files = glob.glob(directory +'**', recursive=True)
    result_dict = {}
    file_count = 0
    for file in files:
        if '.' in file:
            try:
                file_count += 1
                test_image = image.load_img(file, target_size = shape)
                test_image = image.img_to_array(test_image)
                test_image = np.expand_dims(test_image, axis = 0)
                result = classifier.predict(test_image)
                if result[0][0] == 1:
                    result_dict[file.split('/')[-1]] = 'Paul'
                else:
                    result_dict[file.split('/')[-1]] = 'Not Paul'
            except:
                pass
    return result_dict, file_count
```

```python
# Paul holdout
paul_dict, paul_file_count = paul_dict_test(paul)
sum([val == 'Paul' for val in paul_dict.values()])/len(paul_dict), paul_file_count
```

```python
# Cats copy
cat_dict, cat_file_count = paul_dict_test(cats)
sum([val == 'Paul' for val in cat_dict.values()])/len(cat_dict), cat_file_count
```

```python
# Dogs original
dog_dict, dog_file_count = paul_dict_test(dogs)
sum([val == 'Paul' for val in dog_dict.values()])/len(dog_dict), dog_file_count
```

```python
# Actors
actors_dict = {}
for file in glob.glob(actors + '/*', recursive=True):
    result_dict, count = paul_dict_test(file)
    if len(result_dict) > 0:
        percent = sum([val == 'Paul' for val in result_dict.values()])/len(result_dict)
        try:
            actors_dict[file.split('/')[-1]] = percent, count
            print(file.split('/')[-1], percent, count)
        except:
            print(file.split('/')[-1], 'did not work. ')
```

```python
actors_df = pd.DataFrame(actors_dict).T.reset_index()
actors_df.columns = ['Actors','Percent_Paul', 'Count']
actors_df
```
